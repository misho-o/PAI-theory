## 1. CONFIGURABLE ENDPOINTS

```python
try:
  with open('apiKey.txt','r') as file:
    apiKey = file.read()

except OSError as e:
  print(e)
```

```python
import requests
```

## 2. ROBUST ERROR HANDLING

```python
try:
  response = requests.get(f'https://api.cricapi.com/v1/series_info?apikey={apiKey}&offset=0&id=47b54677-34de-4378-9019-154e82b9cc1a')
except requests.exceptions.RequestException as e:
  print(e)
```

```python
data = response.json()
```

```python
data
```

## 3. API STATUS

```python
data['info']
```

## 4. GRACEFUL API FAILURES

```python
if data['status'] == "success" and data.get('data'):
  print("Status Succesfull!")
else:
  print(f'{data['reason']}')
```

## 5. FORMATTED MATCH DISPLAY

```python
matchData = data['data']['matchList']
```

```python

for match in matchData:
  print(f"-------------------------\nName:{match['name']}\nStatus:{match['status']}\nVenue:{match['venue']}\nDate:{match['date']}\n-------------------------")
```

## 6. NESTED SCORE DETAILS

```python
try:
  scoreResponse = requests.get(f'https://api.cricapi.com/v1/match_info?apikey={apiKey}&offset=0&id=820cfd88-3b56-4a6e-9dd8-1203051140da')
  scoreResponse= scoreResponse.json()
except requests.exceptions.RequestException as e:
  print(e)
```

```python
if scoreResponse['status'] == "success" and scoreResponse.get('data'):
  print("Status Succesfull!")
else:
  print(f'{scoreResponse['reason']}')
```

```python
scores = scoreResponse['data']['score']
```

```python
print(f'{scoreResponse['data']['name']}')
for score in scores:
  print(f"-------------------------\nRuns:{score['r']}\nWickets:{score['w']}\nOvers:{score['o']}\nInning:{'inning'}")
```
